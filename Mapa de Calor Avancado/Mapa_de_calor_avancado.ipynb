# Bibliotecas

```python
import folium
import branca.colormap
import numpy as np
from folium.plugins import HeatMap
```

# Dados de entrada

```python
geojson_arquivo = "geojson_municipios_rj.json"
dados = np.genfromtxt("dados_calor.csv", delimiter = ",")
```

# Mapa

```python
mapa_rj = folium.Map([-22.42,-42.48], zoom_start = 7, tiles = None)

#Adicionar o fundo branco
folium.TileLayer(tiles = branca.utilities.image_to_url([[1,1], [1,1]]),
                attr = "Luana Almeida", name = "Imagem Fundo").add_to(mapa_rj)

#Adicionando a fronteira dos municipios
estilo = lambda x: {"color" : "black",
                   "fillOpacity": 0,
                   "weight": 1}

folium.GeoJson(geojson_arquivo, style_function = estilo,
               name = "Municipios").add_to(mapa_rj)

#Paleta de cores
indices = [0, 0.3, 0.7, 1]

colormap = branca.colormap.StepColormap(["green", "yellow", "red"], index = indices,
                                       caption = "Índice Aleatório")

dicionario_cores = {0: "green",
                   0.3: "green",
                   0.301: "yellow",
                   0.7: "yellow",
                   0.701: "red",
                   1: "red"}


colormap.scale(0, 500).add_to(mapa_rj)

#Mapa de Calor
HeatMap(data = dados,
       gradient = dicionario_cores,
       min_opacity = 0.1,
       radius = 20,
       blur = 5,
       name = "Dados Calor").add_to(mapa_rj)


#Controle de camadas
folium.LayerControl(position = "topleft").add_to(mapa_rj)

#Salvando o mapa
mapa_rj.save("mapa_calor_avancado.html")
mapa_rj
```
